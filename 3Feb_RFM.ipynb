#### RFM Model

1) Used for Customer Segmentation

#### RFM

1) <b>R - Recency</b><br>
How many days since the customer has made the last transaction.

3) <b>F - Frequency</b><br>
The number of times a customer made the transaction.

4) <b>M - Monetory</b><br>
The total Amount spent by the customer in his/her lifetime.

#### Recency (Should be minimum)

```python
rec = df.groupby('Customer ID')['Order Date'].max().reset_index()
rec.head()
```

```python
rec['Recency'] = (df['Order Date'].max() - rec['Order Date']).dt.days
rec.head()
```

#### Frequency (Should be Maximum)

```python
freq = df['Customer ID'].value_counts().reset_index()
freq.columns = ['Customer ID','Frequency']
freq.head()
```

```python
print(rec.shape)
print(freq.shape)
```

#### Monetory (Should be maximum)

```python
mon = df.groupby(['Customer ID'])['Sales'].sum().reset_index()
mon.columns = ['Customer ID','Monetory']
mon.head()
```

```python
mon.shape
```

#### RFM

```python
rf = pd.merge(rec,freq, on='Customer ID')
rf.head()
```

```python
rfm = pd.merge(rf,mon,on='Customer ID')
rfm.head()
```

```python
rfm.drop('Order Date',axis=1,inplace=True)
```

```python
rfm.head()
```

```python
rfm1 = rfm.copy()
rfm2 = rfm.copy()
```

#### When data needs to be sorted in different order withwrt same column - 
Based on Customer_ID, sort Recency in asc order, Freq and Monetory in desc order

Quartile Analysis(4 parts), Decile Analysis(10 parts), Pentile Anlysis(5 parts)

#### Pentile Analysis

```python
z1 = rfm.describe(percentiles=[0.20,0.40,0.60,0.80]).T
z1 = z1.iloc[:,3:]
z1
```

```python
rfm['RScore'] = pd.cut(rfm['Recency'],bins=[-1,26,53,103,224.6,1165],labels=[5,4,3,2,1])

rfm['FScore'] = pd.cut(rfm['Frequency'],bins=[0,7,10,13,18,37],labels=[1,2,3,4,5])

rfm['MScore'] = pd.cut(rfm['Monetory'],bins=[4.833, 961.5406,1739.4324,2750.748, 4282.6344, 25043.05]
                                            ,labels=[1,2,3,4,5])
rfm.head()
```

```python
rfm['RF_Score'] = rfm['RScore'].astype(str) + rfm['FScore'].astype(str)

rfm['RFM_Score'] = rfm['RScore'].astype(str) + rfm['FScore'].astype(str) + rfm['MScore'].astype(str)
rfm.head()
```

<img src="rf_chart.png">

```python
rfm['Segment'] = rfm['RF_Score'].replace({
        '[1-2][1-2]' : 'Hibernating',
        '3[1-2]'  : 'About to Sleep',
        '41' : 'Promising',
        '51' : 'New Customers',
        '33' : 'Needs Attention',
        '[1-2][3-4]' : 'At Risk',
        '[1-2]5' : 'Cant Loose them',
        '[4-5][2-3]' : 'Potential Loyalists',
        '[3-4][4-5]' : 'Loyal Customers',
        '5[4-5]' : 'Champions'
},regex=True)
```

```python
rfm['Segment'].value_counts()
```

```python
profile = rfm.groupby(['Segment']).agg({'Segment':'count',
                             'Recency' :'mean','Frequency' :'mean','Monetory' :'mean'})
profile
```
